```python
penguins_df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
print(penguins_df)
```

```python
#Suitable graph of the mean BMI (BMI not body_mass weight_in_kg/height(m2))
 #of female and male penguins. (Assume the height is the flipper length)
# Assuming height is flipper length, calculate BMI for each penguin
penguins_df['bmi'] = penguins_df['body_mass_g'] / ((penguins_df['flipper_length_mm'] / 1000) ** 2)

# Plot the mean BMI for female and male penguins
sns.barplot(x='sex', y='bmi', data=penguins_df, estimator='mean', errorbar=None)
plt.title('Mean BMI of Female and Male Penguins')
plt.xlabel('Sex')
plt.ylabel('Mean BMI')
plt.show()
```

```python
museum_df=pd.read_csv('museum_visitors.csv',delimiter=',')
museum_df.head(12)
```

```python

museum_df['Date'] = pd.to_datetime(museum_df['Date']) # Convert the 'Date' column to a datetime object because
                                                       # it's currently a mix of string and numbers
museum_df.set_index('Date', inplace=True)
museum_df.plot()
plt.title('Number of visitors to each museum over time')
plt.xlabel('Date')
plt.ylabel('Number of visitors')
plt.show()
```

```python
gas_df=pd.read_csv('gas_prices.csv',delimiter=',')
# Transpose the gas_df dataframe to ensure gas_price has the same number of elements as countries
#heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming gas_df is your DataFrame and it's already been imported

# Define the color scale range
vmin = gas_df.min() # This sets the minimum to the lowest gas price in your dataset
vmax = gas_df.max()  # This sets the maximum to the highest gas price in your dataset
print(vmin)
print(vmax)
# Optionally, you can set specific values for vmin and vmax
# For example:
# vmin = 0.5
# vmax = 5.0

# Create the heatmap with a specific range for the color scale
plt.figure(figsize=(12, 8))
sns.heatmap(gas_df, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, vmin=0.5, vmax=8)
plt.title('Gas Prices by Country Over Years')
plt.ylabel('Year')
plt.xlabel('Country')

# Show the plot
plt.show()





```
