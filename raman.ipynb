#### Task: Apply Data Tidying techniques to get a tidier dataset with more representive columns, please do it step-by-step and explain each step you are doing.

```python
import pandas as pd

# Load the raw dataset
df = pd.read_csv("data/tb-raw.csv")
```

```python
print(df.head())
print(df.info())
```

```python
sex_age_columns = [col for col in df.columns if col.startswith(('m', 'f'))]
for col in sex_age_columns:
    sex = col[0]  # Extracting sex from column name
    age_group = col[1:]  # Extracting age group from column name
    df[f'{sex}_sex_{age_group}_age'] = df[col]
```

```python
df.drop(columns=sex_age_columns, inplace=True)
```

```python
import pandas as pd

# Replace "your_dataset_url_here" with the actual URL or file path
url = "http://lib.stat.cmu.edu/datasets/boston"
cols = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

# Read the dataset with corrected parameters
boston = pd.read_csv(url, sep="\s+", skiprows=21, header=None, names=cols)

# Display the first few rows of the dataset
print(boston.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'boston' is the DataFrame containing the Boston housing dataset
# Make sure to replace 'boston' with the actual name of your DataFrame

# Check the columns of the DataFrame
print(boston.columns)

# Visualizing the distribution of the target variable 'MEDV' (assuming it's the target variable in your Boston dataset)
plt.figure(figsize=(10, 6))
sns.histplot(boston['MEDV'], kde=True, bins=30)
plt.title('Distribution of Housing Prices')
plt.xlabel('Price ($1000s)')
plt.ylabel('Frequency')
plt.show()


```

```python
# Scatter plot of 'MEDV' vs 'RM' (average number of rooms per dwelling)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RM', y='MEDV', data=boston)
plt.title('Average Number of Rooms vs Price')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Price ($1000s)')
plt.show()

# Scatter plot of 'MEDV' vs 'LSTAT' (% lower status of the population)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LSTAT', y='MEDV', data=boston)
plt.title('% Lower Status of the Population vs Price')
plt.xlabel('% Lower Status of the Population')
plt.ylabel('Price ($1000s)')
plt.show()


```

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Z-score outlier detection for 'MEDV' column
z_scores = np.abs(stats.zscore(boston['MEDV']))
threshold = 3
outliers_z = np.where(z_scores > threshold)[0]

# Display the z-score outliers
print("Z-score outliers:")
print(boston.loc[outliers_z, ['RM', 'LSTAT', 'MEDV']])

# Scatter plot without outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RM', y='MEDV', data=boston.drop(outliers_z))
plt.title('Average Number of Rooms vs Price (without outliers)')
plt.xlabel('Average Number of Rooms')
plt.ylabel('Price ($1000s)')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='LSTAT', y='MEDV', data=boston.drop(outliers_z))
plt.title('% Lower Status of the Population vs Price (without outliers)')
plt.xlabel('% Lower Status of the Population')
plt.ylabel('Price ($1000s)')
plt.show()

```

```python
z_scores = np.abs(stats.zscore(boston['MEDV']))
threshold = 3
outliers_z = np.where(z_scores > threshold)[0]
print("Number of outliers detected using z-score method:", len(outliers_z))

# IQR outlier detection for 'MEDV' column
Q1 = boston['MEDV'].quantile(0.25)
Q3 = boston['MEDV'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = boston[(boston['MEDV'] < lower_bound) | (boston['MEDV'] > upper_bound)]
print("Number of outliers detected using IQR method:", len(outliers_iqr))
```

```python
# Removing outliers using IQR method
boston_no_outliers = boston[~((boston['MEDV'] < lower_bound) | (boston['MEDV'] > upper_bound))]

# Visualizing the distribution of the target variable after removing outliers
plt.figure(figsize=(10, 6))
sns.histplot(boston_no_outliers['MEDV'], kde=True, bins=30)
plt.title('Distribution of Housing Prices (No Outliers)')
plt.xlabel('Price ($1000s)')
plt.ylabel('Frequency')
plt.show()
```
